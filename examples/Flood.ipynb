# Simple example

## Init true and prediction labelled data

```python
true = pd.Series(0,pd.date_range('2020-01-01','2020-01-10',freq='D'))
true.iloc[4]=1


prediction = pd.Series(0,pd.date_range('2020-01-01','2020-01-10',freq='D'))
prediction.iloc[[5,6,7]]=1


pd.concat([true,prediction],1).reset_index()
```

## Evaluating by default `NAB` metric

```python
results = evaluating(true=[[pd.Timestamp('2020-01-05')],[pd.Timestamp('2020-01-05')],[pd.Timestamp('2020-01-05')]],prediction=[prediction.copy(),prediction.copy(),prediction], metric='nab',numenta_time='3D',
                     anomaly_window_destenation='center',
                     clear_anomalies_mode=True,plot_figure=True)
print(results)
```
